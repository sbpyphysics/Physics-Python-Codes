The above example is very very use full to understand how to use array and def and then from there how Can we plot a function, the function is generated from a simple looking function given by $z^2 - 1$ where z is a complex variable.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable

def f(z):
    return z - (2 * np.power(np.tan(z), 2) / (np.sin(z) * np.cos(z)))



x = np.linspace(-2,2,1500)
y = np.linspace(-2,2,1500)
xm,ym = np.meshgrid(x,y)
mesh = xm + 1j*ym
def divertr(mesh, nmiter=50, radius=3.0):
    z = mesh.copy()
    diverlen = np.zeros(mesh.shape) #Creates an matrix of same mesh shape...
    for i in range(nmiter):
        mask = np.abs(z)<2 #Checking the mesh grid points if it is under 2.0 or not
        diverlen[mask] += 1 #If it is under 2.0 it increases and store it in the diverlen zeros array
        z[mask] = f(z[mask]) #it calculates the function for all the bounded points simultaneously for 10 times
    return diverlen

output = divertr(mesh)
fig = plt.figure(figsize=(12, 8))
ax = plt.axes()

ax.set_xlabel('Real axis')
ax.set_ylabel('Imaginary axis')
im = ax.imshow(output, extent=[-2, 2, -2, 2], cmap = 'GnBu')
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.85)
plt.colorbar(im, cax=cax, label='Number of iterations');


















```

Let us take an example of the FDM process then explain the theory later...
let us consider we have a problem where $d^2y/dx^2 = -B$ where B is some constant for the sake of problem we can assign any numerical value to it.


```python
import numpy as np
import matplotlib.pyplot as plt
#Let us define the matrix first from the Theory of FDM
n = 15 #Any arbitrary number of steps we assign
A = np.zeros((n+1,n+1),dtype = float)
h = (15-12)/n+1 #Step Size
B = 5
#The diagonal elements are mostly -2 except the first and last which come from the boundary condition.
A[0,0] = 1
A[n,n] = 1
for i in range(1,n): #This is How we put the values or construct the matrix where we eliminate the first and last two elements
    A[i,i] = -2
for k in range(1,n):
    A[k,k-1] = 1
    A[k,k+1] = 1

print(A)
c = -B*np.square(h)

rhs = c* np.ones(n+1)
rhs[0] = 12
rhs[15] = 15

print(rhs)


solve = np.linalg.solve(A,rhs)
print(solve)
print(solve.shape)
x = np.linspace(0,15,n+1)
plt.plot(x,solve,marker = 'o')
plt.grid(True)


```

### Now lets see some example of enumerate...
If we have a Numpy array, and we are iterating it, and we also need its position element wise, in that case enumerate gives us a convinient way to iterate it and also indexing at the same time.


```python
import numpy as np
my_array = np.array([10, 20, 30, 40, 50]) #This is a 1d array
for index, value in enumerate(my_array): #We need both index and the corresponding value at the same time so we are getting that
  print(f"Index: {index}, Value: {value}")
```

```python
#Also if we want to start from some arbitrary Index point that is also possible.

import numpy as np
my_array = np.array([10, 20, 30, 40, 50,60,70,80]) #This is a 1d array
for index, value in enumerate(my_array,start = 2): #We need both index and the corresponding value at the same time so we are getting that, we just need to include start option and assign a value...
  print(f"Index: {index}, Value: {value}")
```
